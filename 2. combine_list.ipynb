```python
import os
import pandas as pd
```

```python
current_path = os.getcwd()
raw_list_path = os.path.join(current_path, 'Raw List')
raw_list = os.listdir(raw_list_path)
```

```python
df_combined = pd.DataFrame()
for i in raw_list:
    if i.endswith('.csv'):
        df = pd.read_csv(os.path.join(raw_list_path, i))
        df_combined = pd.concat([df_combined, df], axis=0)
    
```

```python
df_combined.reset_index(drop=True, inplace=True)
```

```python
df_combined = df_combined.drop(['Unnamed: 0'], axis=1)
```
